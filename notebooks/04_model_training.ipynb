# Intelligent Login Risk Analyzer  
## Notebook 04: Model Training

This notebook trains an anomaly detection model using the engineered features prepared in the previous notebook.

We use an unsupervised learning approach because real anomaly labels are usually limited in cybersecurity scenarios.

## Step 1: Import Required Libraries

We import the libraries required for loading the processed dataset and training the anomaly detection model.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
```

## Step 2: Load Processed Feature Dataset

We load the feature-engineered dataset created in Notebook 03.

```python
X = pd.read_csv('../outputs/processed_features.csv')
X.head()
```

## Step 3: Review Feature Dataset Shape

This step confirms the number of rows and columns available for model training.

```python
print("Feature dataset shape:", X.shape)
```

## Step 4: Initialize the Isolation Forest Model

Isolation Forest is an unsupervised anomaly detection algorithm that isolates unusual observations instead of learning from labeled attack data.

It is suitable for this project because:
- anomaly labels are not fully available
- suspicious login behavior is rare compared to normal behavior
- it works well for outlier detection tasks

```python
model = IsolationForest(
    n_estimators=100,
    contamination=0.1,
    random_state=42
)
```

## Step 5: Train the Model

We fit the Isolation Forest model on the processed login features.

```python
model.fit(X)
```

## Step 6: Generate Anomaly Predictions

The model returns:
- `1` for normal observations
- `-1` for anomalies

We store the predictions for further analysis.

```python
predictions = model.predict(X)
predictions[:10]
```

## Step 7: Add Predictions to the Dataset

We attach the model predictions to the feature dataset for interpretation.

```python
X['anomaly_prediction'] = predictions
X.head()
```

## Step 8: Count Normal vs Anomalous Records

This step shows how many records were classified as normal and how many were flagged as anomalies.

```python
X['anomaly_prediction'].value_counts()
```

## Step 9: Convert Predictions into Readable Labels

For better interpretation, we map the prediction values into human-readable classes.

```python
X['anomaly_label'] = X['anomaly_prediction'].map({1: 'Normal', -1: 'Anomaly'})
X[['anomaly_prediction', 'anomaly_label']].head()
```

## Step 10: Inspect Detected Anomalies

We filter the records predicted as anomalies to understand what the model is flagging.

```python
anomalies = X[X['anomaly_label'] == 'Anomaly']
anomalies.head(10)
```

## Additional Note

The anomaly predictions are model-based outlier signals, not confirmed attacks.

These outputs should be interpreted as suspicious records requiring further review, not as final proof of compromise.

## Step 11: Save Model Predictions

We save the prediction results for use in the evaluation notebook.

```python
X.to_csv('../outputs/model_predictions.csv', index=False)
```

## Summary

In this notebook, we:
- loaded the processed feature dataset
- trained an Isolation Forest anomaly detection model
- generated anomaly predictions
- labeled records as normal or anomalous
- saved prediction results for further evaluation

This notebook establishes the first machine learning layer of the Intelligent Login Risk Analyzer.
